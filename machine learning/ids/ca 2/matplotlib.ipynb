```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
noOfWeekdays = [1,2,3,4,5,6,7]
temperature = [35,37,40,29,30,31,33]
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(noOfWeekdays, temperature, color='green', linewidth=2, linestyle='dotted')
```

```python
plt.plot(noOfWeekdays, temperature, 'rd')
```

```python
plt.plot(noOfWeekdays, temperature, color='red', marker='*', linestyle='dashed', alpha=0.5)
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,49,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t,label='Max')
plt.plot(days, min_t, label='Mint')
plt.plot(days, avg_t, label='Avg')

plt.legend(loc="best", shadow=True, fontsize='small')
plt.grid()
```

```python
company = ['GOOGL','AMZN','MSFT','META']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
plt.ylabel('Revenue is $USD')
plt.title('US Tech Stocks')
plt.bar(company, revenue, label='Revenue')
plt.bar(company, profit, label='Profit')  # Assuming profit is aligned with companies
plt.legend()
plt.show()
```

```python
bloogSugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
bloodSugarWomen = [67,98,44,120,135,150,84,69,89,79,120,112,100]
plt.xlabel('Blood Sugar Range')
plt.ylabel('Total number of Patients')
plt.title('Bloog Sugar Analysis')
plt.hist([bloogSugar,bloodSugarWomen], bins=[80,100,125,150], rwidth=0.99, color=['green','orange'] , label=['men', 'women']) #histtype='step'
plt.legend()
```
